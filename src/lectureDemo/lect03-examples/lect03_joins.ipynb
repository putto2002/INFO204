# Joining Data Frames using pandas
These examples are taken straight from the [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html)! :)

```python
import pandas as pd
```

By default, we can stack data frames using the concat() function

```python
df1 = pd.DataFrame({"A": ["A0", "A1", "A2", "A3"],
                    "B": ["B0", "B1", "B2", "B3"],
                    "C": ["C0", "C1", "C2", "C3"],
                    "D": ["D0", "D1", "D2", "D3"],
                   }, index=[0, 1, 2, 3]
)

df2 = pd.DataFrame({"A": ["A4", "A5", "A6", "A7"],
                    "B": ["B4", "B5", "B6", "B7"],
                    "C": ["C4", "C5", "C6", "C7"],
                    "D": ["D4", "D5", "D6", "D7"],
                   }, index=[4, 5, 6, 7],
)

df3 = pd.DataFrame({"A": ["A8", "A9", "A10", "A11"],
                    "B": ["B8", "B9", "B10", "B11"],
                    "C": ["C8", "C9", "C10", "C11"],
                    "D": ["D8", "D9", "D10", "D11"],
                    }, index=[8, 9, 10, 11],
)

frames = [df1, df2, df3]

result = pd.concat(frames)
result
```

We can change the stacking behaviour to stack horizontally by changing the axis argument (which will then try to match on common row indices):

```python
pd.concat(frames, axis=1)
```

That didn't seem too useful, but serves as the basis for relational joins (i.e., SQL-like joins). Here's an example of an outer join (the default):

```python
 df4 = pd.DataFrame({"B": ["B2", "B3", "B6", "B7"],
                     "D": ["D2", "D3", "D6", "D7"],
                     "F": ["F2", "F3", "F6", "F7"],
                    }, index=[2, 3, 6, 7],
)
result = pd.concat([df1, df4], axis=1)
result
```

And the same operation, only this time requiring an inner join logic:

```python
pd.concat([df1, df4], axis=1, join='inner')
```
