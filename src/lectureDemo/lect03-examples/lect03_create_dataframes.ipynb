# Creating pandas DataFrame objects

pandas allows you to create data frames manually, via combining a selection of Series objects, or allows you to import a range of different file types (e.g., CSV, Excel, fwf, tabbed, ...)

Let's look at creating a data frame by hand, by using a dictionary of pre-defined Series objects:

```python
import pandas as pd

s1 = pd.Series([1, 2, 3, 4])
s2 = pd.Series(['a', 'b', 'c', 'd'])

df = pd.DataFrame({ 'a': s1, 'b': s2 })
df
```

That was pretty straightforward. Now, lets see how we can load in data from an existing source (in this case a CSV file):

```python
df = pd.read_csv('../data/weights.csv')
df
```

Again, pretty straightforward!
